```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#data collection and preprocessing
bank_authentication=pd.read_csv("/content/BankNote_Authentication.csv")
```

```python
print(bank_authentication.head(5))
```

```python
X=bank_authentication.drop(columns='class',axis=1)
Y=bank_authentication['class']
```

```python
print(X,Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
print(X_train,Y_train)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_predictions=model.predict(X_train)
X_train_accuracy=accuracy_score(X_train_predictions,Y_train)
print(X_train_accuracy)
```

```python
X_test_predictions=model.predict(X_test)
X_test_accuracy=accuracy_score(X_test_predictions,Y_test)
print(X_test_accuracy)
```

```python
input_data=(-3.6227,3.9958,-0.35845,-3.9047)
input_data_as_nparray=np.asarray(input_data)
input_data_reshaped=input_data_as_nparray.reshape(1,-1)
predictions=model.predict(input_data_reshaped)
print(predictions)
if(predictions==0):
  print("forged i.e fake")
else:
  print("genuine i.e real")
```
